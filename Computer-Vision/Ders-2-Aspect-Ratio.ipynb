```python
import cv2
def resizewithAspectRatio(img,width=None,height=None,inter=cv2.INTER_AREA):
    dimension = None
    (h,w) = img.shape[:2]
    if width is None and height is None:
        return img
    if width is None:
        ratio = height/h
        img = cv2.resize(img,(height,int(ratio*w)))
    elif height is None:
        ratio = width/w
        img = cv2.resize(img,(int(h*ratio),width))
    else:
        img = cv2.resize(img,(height,width))
    return img
```

```python
img =cv2.imread("klon1.png")
img = resizewithAspectRatio(img,width=300,height=None,inter=cv2.INTER_AREA)
#cv2.namedWindow("image",cv2.WINDOW_NORMAL)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
