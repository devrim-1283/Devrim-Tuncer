```python
import cv2
import numpy as np
```

```python
canvas = np.zeros((512,512,3),dtype="uint8")
print(canvas)

cv2.imshow("Image",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
canvas = np.zeros((512,512,3)) + 255
print(canvas)
cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
  # Line çizimi yapma
canvas = np.zeros((512,512,3)) + 255
print(canvas)
cv2.line(canvas,(0,0),(512,512),(0,0,0), thickness=5)
cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Üçgen oluşturma özel bir fonksiyon yok biz çizgilerle oluşturabiliriz
canvas = np.zeros((512,512,3)) + 255

cv2.line(canvas,(50,50),(150,150),(255,0,0),thickness=3)
cv2.line(canvas,(150,150),(150,80),(255,0,0),thickness=3)
cv2.line(canvas,(150,80),(50,50),(255,0,0),thickness=3)


cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
canvas = np.zeros((312,312,3)) + 255

cv2.line(canvas,(100,100),(254,259),(0,0,0),thickness=5)
cv2.line(canvas,(254,259),(120,4),(0,0,0,),thickness=5)
cv2.line(canvas,(120,4),(100,100),(0,0,0),thickness=5)

cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# çok kenarlı gen oluşturma

canvas = np.zeros((512,512,3)) + 255

points = np.array([[110,200],[400,500],[124,180],[400,400]],np.int32)


cv2.polylines(canvas,[points],True,(0,0,255),thickness=5)  # true değeri kapalı yapar 

cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# ellipse oluşturma
canvas = np.zeros((512,512)) + 255

cv2.ellipse(canvas, (250,250),(100,150),20,0,360,(0,0,0),thickness=5)
# ilk değer merkez ikinci değer genişil yükseklik ,eğim , başlangıç açısı, bitiş açısı , renk , kalınlık

cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Yazı Yazdırma
canvas = np.zeros((512,512)) + 255

font1 = cv2.FONT_HERSHEY_COMPLEX
font2 = cv2.FONT_HERSHEY_DUPLEX
font3 = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX


cv2.putText(canvas,"Merhaba",(120,120),font1,3,(0,0,0),cv2.LINE_AA)
# ilk kordinat metnin başlayacağı sol alt kordinat
# yazı tipi seçiyorsun



cv2.imshow("Canvas",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
