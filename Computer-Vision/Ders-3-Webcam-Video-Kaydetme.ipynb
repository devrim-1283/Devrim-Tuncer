```python
import cv2
```

```python
cap = cv2.VideoCapture(0)   # webcami tanıttık

while True:
    ret, frame =  cap.read() 

    cv2.imshow("Webcam Live",frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):  # q ya basınca çık
        break

cap.release() # kamerayı takip etme artık
cv2.destroyAllWindows() # pencereyi kapat
```

```python
cap = cv2.VideoCapture(0)   # webcami tanıttık
# hata alırsak o ın yanına cv2.CAP_DSHOW yazıp flag değeri gir
codec = cv2.VideoWriter_fourcc("W","M","V","2")
framerate= 30
resolution = (640,480)
videofileoutput = cv2.VideoWriter("webcam.avi",codec,framerate,resolution)
while True:
    ret, frame =  cap.read() 
    frame = cv2.flip(frame,1) # y eksenine göre yansıt bana frami kafam sola çevrilince ekranda da sola değil sağa çevrilsin
    
    videofileoutput.write(frame) # frame frame kaydet
    
    cv2.imshow("Webcam Live",frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):  # q ya basınca çık
        break

videofileoutput.release() # video kaydetme içini bırak
cap.release() # kamerayı takip etme artık
cv2.destroyAllWindows() # pencereyi kapat
```

```python
# apiden webcam görüntüsü çekerek video kaydı alma

import cv2
import numpy as np
import requests
url = "http://172.16.3.203:8080/shot.jpg"
while True:
    img_resp = requests.get(url)
    img_arr = np.array(bytearray(img_resp.content),dtype=np.uint8)
    img = cv2.imdecode(img_arr,cv2.IMREAD_COLOR)
    img = cv2.resize(img,(640,480))

    cv2.imshow("Android Camera",img)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cv2.destroyAllWindows()
    
```
