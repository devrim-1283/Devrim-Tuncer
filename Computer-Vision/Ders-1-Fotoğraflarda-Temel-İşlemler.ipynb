```python
import cv2
img = cv2.imread("devrim.jpeg") # Resmimizi bir değişkene atıyoruz.
# imshow ilk parametre pencere adı ikinci parametre resmi isstiyor.
cv2.imshow("Image",img)  # Resmimizi pencerede göstermemize yarar fakat alttaki değerleri girmezsek resmi düzgün görüntüleyemeyebiliriz.
cv2.waitKey(0) # Resmin ne kadar milisaniye ekranda duracağı 0 yaparsak biz kapatanaa kadar durur
cv2.destroyAllWindows() # Bu komut önemli bazı hatalar vesayre aldığımızda pencereyi kapatmak için default "q" tuşu ister basınca pencere kapanır.count
# wait key ve destroy komutları kemik kalıp komutlardır.
```

```python
import cv2
img = cv2.imread("devrim.jpeg") # Resmi değişkene attık
# Farkettiyseniz resim boyutunu değiştiremiyoruz pencereden onun için de bir komutumuz var.
cv2.namedWindow("image2",cv2.WINDOW_NORMAL) # İlk parametre pencere ismi ikinci parametrede windowun boyutunu değiştirilebilir yapıyoruz
cv2.imshow("image2",img) # pencere isimleri çakışmasın.
cv2.waitKey(0) # pencerenin kaç saniyeq kalacağını gösterir.
cv2.destroyAllWindows()
cv2.imwrite("klon1.png",img) # resim kaydetme
```

```python
import cv2
img = cv2.imread("devrim.jpeg",0) # Resmi gri tonlarda okumak için 0 parametresi veririz
cv2.namedWindow("image2",cv2.WINDOW_NORMAL) 
cv2.imshow("image2",img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
