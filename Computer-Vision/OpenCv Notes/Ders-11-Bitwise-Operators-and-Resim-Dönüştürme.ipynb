```python
import cv2
import numpy as np
```

```python
dark = np.zeros((512,512),dtype="uint8") 
white = np.zeros((512,512),dtype="uint8") + 255

cv2.circle(dark,(128,256),70,(255,255,255),-1)
cv2.rectangle(dark,(256,0),(512,512),(255,255,255),-1)
cv2.rectangle(white,(256,0),(512,512),(0,0,0),-1)


cv2.imwrite("test_image/dark.png",dark)
cv2.imwrite("test_image/white.png",white)

cv2.imshow("IMG-1",dark)
cv2.imshow("IMG-2",white)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# and Operatr   beyaz == 1  siyah == 0
import cv2 

dark = cv2.imread("test_image/dark.png")
white = cv2.imread("test_image/white.png")


bitwise_and = cv2.bitwise_and(dark,white)

cv2.imshow("And",bitwise_and)
cv2.imshow("Dark",dark)
cv2.imshow("White",white)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# or bitwise operatr
dark = cv2.imread("test_image/dark.png")
white = cv2.imread("test_image/white.png")

bitwise_or = cv2.bitwise_or(dark,white)

cv2.imshow("OR",bitwise_or)
cv2.imshow("Dark",dark)
cv2.imshow("White",white)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# xor operatr    0 0 1 1 ise 0  10 01 ise 1

dark = cv2.imread("test_image/dark.png")
white = cv2.imread("test_image/white.png")

bitwise_xor = cv2.bitwise_xor(dark,white)

cv2.imshow("XOR",bitwise_xor)
cv2.imshow("Dark",dark)
cv2.imshow("White",white)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# not operatr 

dark = cv2.imread("test_image/dark.png")
white = cv2.imread("test_image/white.png")


not_dark = cv2.bitwise_not(dark)
not_white = cv2.bitwise_not(white)


cv2.imshow("Dark",dark)
cv2.imshow("Not Dark",not_dark)

cv2.imshow("White",white)
cv2.imshow("Not White",not_white)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
