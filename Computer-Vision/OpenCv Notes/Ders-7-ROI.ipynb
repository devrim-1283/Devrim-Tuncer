```python
path = "test_image/opencv.png"
img = cv2.imread(path)

print(img.shape)
(b,g,r) = cv2.split(img)
cv2.imshow("Image",img)
cv2.imshow("B",b)
cv2.imshow("G",g)
cv2.imshow("R",r)

merged = cv2.merge([b,g,r])
cv2.imshow("Merged",merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
path = "test_image/opencv.png"
img = cv2.imread(path)
(b,g,r) = cv2.split(img)

black = np.zeros(img.shape[:2],dtype="uint8")

merged_r = cv2.merge([black,black,r])
merged_b = cv2.merge([b,black,black])
merged_g = cv2.merge([black,g,black])
cv2.imshow("Merge Red",merged_r) # sadece ekrandaki kırmızı değerleri göstermiş olduk
cv2.imshow("Merge Blue",merged_b) # sadece mavi değerleri
cv2.imshow("Merge Green",merged_g) # sadece yeşil değerleri
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
