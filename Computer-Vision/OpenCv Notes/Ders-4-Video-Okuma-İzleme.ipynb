```python
cap = cv2.VideoCapture("webcam.avi")

while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame,1)
    
    cv2.imshow("web.avi",frame)
    if cv2.waitKey(1) & 0xFF == ord("q"): # klavyeden q girişini yakala ve döngüye git
        break
cap.release()
cv2.destroyAllWindows()
# görüntünün hızlı olmasının sebebi video karesinin senin karenden yavaş olması waitkey değerini arttıralım
```

```python
import cv2
cap = cv2.VideoCapture("webcam.avi")

while True:
    ret, frame = cap.read()
    if ret==False: # ret==0 da diyebiliriz
        break
    frame = cv2.flip(frame,1)
    
    cv2.imshow("web.avi",frame)
    if cv2.waitKey(10) & 0xFF == ord("q"): # klavyeden q girişini yakala ve döngüye git
        break
cap.release()
cv2.destroyAllWindows()
```
