```python
import cv2
import numpy as np
```

```python
dark = np.zeros((512,512),dtype="uint8") 
white = np.zeros((512,512),dtype="uint8") + 255

cv2.circle(dark,(128,256),70,(255,255,255),-1)
cv2.rectangle(dark,(256,0),(512,512),(255,255,255),-1)
cv2.rectangle(white,(256,0),(512,512),(0,0,0),-1)


cv2.imwrite("test_image/dark.png",dark)
cv2.imwrite("test_image/white.png",white)

cv2.imshow("IMG-1",dark)
cv2.imshow("IMG-2",white)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
