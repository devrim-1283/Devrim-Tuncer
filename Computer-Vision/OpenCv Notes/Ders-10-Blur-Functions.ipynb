```python
import cv2
import numpy as np
```

```python
img = cv2.imread("test_image/foto1.jpeg")

blur = cv2.blur(img,(17,17)) # Pozitif tek sayılar olmalı
cv2.imshow("Img",img)
cv2.imshow("Img Blur",blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("test_image/foto1.jpeg")

blur_g = cv2.GaussianBlur(img,(11,11),cv2.BORDER_DEFAULT)

cv2.imshow("Image",img)
cv2.imshow("Gaussian Blur",blur_g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("test_image/foto1.jpeg")

blur_m = cv2.medianBlur(img,5) 



cv2.imshow("Image",img)
cv2.imshow("Median Blur",blur_m)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("test_image/foto1.jpeg")
"""
Özetle:

d = 9, filtre çapını belirler (ne kadar geniş alana bakacağını).
sigmaColor = 95, benzer renkleri ne kadar koruyacağını ayarlar.
sigmaSpace = 95, uzak piksellerin ne kadar etkili olacağını 
    belirler.

Bu filtre, kenarları koruyarak bulanıklaştırma yapmak için 
kullanılır. Gürültüyü azaltırken keskin kenarları bozmamak 
için etkilidir. Özellikle kenar koruma gereken görüntü 
işleme uygulamalarında (örneğin yüz bulanıklaştırma veya
 karikatür efekti) tercih edilir.

"""

bilateral_filter = cv2.bilateralFilter(img,9,95,95)


cv2.imshow("Image",img)
cv2.imshow("Bilateral Filter",bilateral_filter)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
