```python
def anafonksiyon(islem_adı):

    def toplama(*a):
        toplam = 0
        for i in a:
            toplam+=i
        return toplam

    
    def çarpım(*a):
        carpım=1
        for i in a:
            carpım*=i
        return carpım

    if islem_adı == "toplama":
        return toplama
    else:
        return çarpım
```

```python
fonk = anafonksiyon("toplama")
fonk2 = anafonksiyon("çarpma")
```

```python
fonk(1,2,3,4,5,6,7,8)
```

```python
fonk2(1,2,3,4,5,6,7,8)
```

### DECORATOR

```python

def kareleri_hesapla(liste):
    result_list=list()
    for i in liste:
        result_list.append(i**2)

    a=0
    for i in result_list:
        a+=1
    print(a)


    return result_list



def karekök_hesapla(liste):
    result_list = list()
    for i in liste:
        result_list.append(i**0.5)

    a=0
    for i in result_list:
        a+=1
    print(a)


    return result_list

kareleri_hesapla(range(1,1000000))
karekök_hesapla(range(1,1000000))
```

```python
# böyle 100 fonksiyon olduğunu düşünün hepsine de ortak bir özellik eklemek istiyoruz ne yapcaz bakalım


def decorator(function_name):
    def işlem(liste_of_func):

        result = function_name(liste_of_func)

        a=0
        for i in result:
            a+=1
        print(a)

        return result
    return işlem


@decorator
def kareleri_hesapla(liste):
    result_list=list()
    for i in liste:
        result_list.append(i**2)
    return result_list


@decorator
def karekök_hesapla(liste):
    result_list = list()
    for i in liste:
        result_list.append(i**0.5)
    return result_list

kareleri_hesapla(range(1,1000000))
karekök_hesapla(range(1,1000000))
        
```
